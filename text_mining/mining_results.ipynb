```python
import json
import matplotlib.pyplot as plt
from glob import glob
from collections import Counter
```

```python
json_files = glob("results/*.json")
```

```python
mining_results = []
for p in json_files:
    with open(p) as f:
        mining_results.append(json.load(f))
```

```python
keywords = ["snakemake", "nextflow", "cromwell", "WDL", "synthetic"]
summary = {
    'abstract': {
        kw : Counter() for kw in keywords
    },
    'title': {
        kw : Counter() for kw in keywords
    }
}
```

```python
for r in mining_results:
    for part in ['abstract', 'title']:
        for kw in keywords:
            summary[part][kw] += r[part][kw]
```
